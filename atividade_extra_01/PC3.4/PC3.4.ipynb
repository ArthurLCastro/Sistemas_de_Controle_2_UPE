```python
# Usando control.matlab.lsim

import numpy as np

x0 = [0, -1, 1]
t = np.arange(0, 10, 0.1)
u = 0*t

yout, T, xout = lsim(
    sys=sys_ss,
    U=u,
    T=t,
    X0=x0
)
```

```python
plt.figure(1)
plt.plot(T, yout)
plt.title('Resposta do sistema à condição inicial')
plt.xlabel('Tempo (s)')
plt.ylabel('Saída y(t)')
plt.grid(True)

plt.figure(2)
plt.plot(T, xout)
plt.title('Respostas individuais das variáveis de estado do sistema')
plt.xlabel('Tempo (s)')
plt.ylabel('Estados x(t)')
plt.grid(True)
plt.legend(['x1', 'x2', 'x3'])

plt.show(block=True)
```
