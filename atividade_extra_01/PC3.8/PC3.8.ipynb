## **Resolução da PC3.8** *(Incompleta)*

```python
from control.matlab import *
import matplotlib.pyplot as plt

B = [
    [0],
    [0],
    [1],
]

C = [
    [1, 0, 0]
]

D = [
    [0]
]
```

```python
for k in range(101):

    A = [
        [0, 1, 0],
        [0, 0, 1],
        [-2, -k, -2],
    ]

    sys_ss = ss(A, B, C, D)

    yout, T = impulse(
        sys=sys_ss,
        T=15,
    )

    plt.plot(T, yout, label=f'K={k}')

plt.title('Respostas ao impulso para cada valor de k')
plt.xlabel('Tempo (s)')
plt.ylabel('Saídas y(t)')
plt.grid(True)
plt.legend()
plt.show(block=True)
```
