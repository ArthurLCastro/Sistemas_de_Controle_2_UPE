## **Resolução da PC3.7**

```python
from control.matlab import *

A = [
    [0, 1],
    [-2, -3],
]

B = [
    [0],
    [1],
]

C = [
    [1, 0]
]

D = [
    [0]
]
```

```python
sys_ss = ss(A, B, C, D)
print(sys_ss)
```

```python
import numpy as np

x0 = [1, 0]
t = np.arange(0, 10, 0.1)
u = 0*t

yout, T, xout = lsim(
    sys=sys_ss,
    U=u,
    T=t,
    X0=x0
)
```

```python
import matplotlib.pyplot as plt

plt.figure(1)
plt.plot(T, xout)
plt.title('Respostas individuais das variáveis de estado do sistema')
plt.xlabel('Tempo (s)')
plt.ylabel('Estados x(t)')
plt.grid(True)
plt.legend(['x1', 'x2'])

plt.show(block=True)
```
