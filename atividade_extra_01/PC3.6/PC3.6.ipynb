## **Resolução da PC3.6**

```python
from control.matlab import *
```

**Resolução da Letra a**

```python
controlador_sys_tf = tf([3], [1, 3])
print(controlador_sys_tf)
controlador_sys_ss = ss(controlador_sys_tf)
print(controlador_sys_ss)
```

**Resolução da Letra b**

```python
processo_sys_tf = tf([1], [1, 2, 5])
print(processo_sys_tf)
processo_sys_ss = ss(processo_sys_tf)
print(processo_sys_ss)
```

**Resolução da Letra c**

```python
sys_s = series(controlador_sys_ss,processo_sys_ss)
sys_cl = feedback(sys_s,[1])    # [1] para feedback positivo
yout, T = impulse(sys_cl)
```

```python
import matplotlib.pyplot as plt

plt.figure(1)
plt.plot(T, yout)
plt.title('Resposta ao impulso')
plt.xlabel('Tempo (s)')
plt.ylabel('Saída y(t)')
plt.grid(True)
```
